```python
import openpyxl

# Carica il file Excel
file = r'..\MIDEA\base spareorder MWO MIDEA_PO 2025-2 (all brands).xlsx'  # Cambia con il tuo nome file
wb = openpyxl.load_workbook(file, data_only=True)
ws = wb.active

results = []

# Trova l'ultima colonna con dati nella prima riga (dalla colonna I in poi)
start_col = 9  # I = colonna 9 (1-based)
max_col = ws.max_column

# Prendi i valori di intestazione
header_row = 1
anagrafica_row = 54  # Riga 54 (1-based, quindi 54)
start_data_row = 2
end_data_row = 50

for col in range(start_col, max_col + 1):
    header_value = ws.cell(row=header_row, column=col).value
    anagrafica_value = ws.cell(row=anagrafica_row, column=col).value

    # Skippa se nessuna intestazione/valore anagrafico (opzionale)
    if header_value is None and anagrafica_value is None:
        continue

    # Vai sulle righe dati
    for row in range(start_data_row, end_data_row + 1):
        cell_value = ws.cell(row=row, column=col).value

        if cell_value is not None and str(cell_value).strip() != "" and str(cell_value).strip() != "0":
            # Prendi la data corrispondente nella colonna D (colonna 4)
            date_value = ws.cell(row=row, column=3).value
            results.append({
                "header": header_value,
                "anagrafica": anagrafica_value,
                "value": cell_value,
                "date": date_value
            })

# Se vuoi salvarlo in un CSV:
import pandas as pd
df = pd.DataFrame(results)
```

```python
import openpyxl
import pandas as pd

file = r'..\HISENSE\base_spareorder HISENSE LVB (all brand) 2025.3.xlsx'  # Cambia nome file se necessario

wb = openpyxl.load_workbook(file, data_only=True)
ws = wb.active

results = []

max_row = ws.max_row
max_col = ws.max_column

start_data_col = 6   # F = colonna 6 (1-based)
start_model_row = 5  # Modelli dalla riga 5 in poi

for row in range(start_model_row, max_row + 1):
    modello = ws.cell(row=row, column=2).value  # Colonna B = 2
    if not modello:
        continue

    for col in range(start_data_col, max_col + 1):
        value = ws.cell(row=row, column=col).value
        if value is not None and str(value).strip() != "" and str(value).strip() != "0":
            header = ws.cell(row=4, column=col).value
            data = ws.cell(row=2, column=col).value  # Data completa in riga 2
            results.append({
                "modello": modello,
                "header": header,
                "data": data,
                "valore": value
            })

# Esporta in CSV
df = pd.DataFrame(results)
```

```python
import openpyxl
import pandas as pd

file = r"..\HOMA\new Generale ordini frigo HOMA in progress_25.xlsx"  # Cambia col nome del tuo file

wb = openpyxl.load_workbook(file, data_only=True)
ws = wb.active

results = []

start_data_col = 8    # H = 8 (1-based)
end_data_col = 92     # CN = 92 (1-based)
start_model_row = 4   # Di solito riga 4, cambia se serve
max_row = ws.max_row

for row in range(start_model_row, max_row + 1):
    modello = ws.cell(row=row, column=2).value  # Colonna B = 2
    if not modello:
        continue

    for col in range(start_data_col, end_data_col + 1):
        value = ws.cell(row=row, column=col).value
        if value is not None and str(value).strip() != "" and str(value).strip() != "0":
            giorno_mese = ws.cell(row=3, column=col).value
            anno = ws.cell(row=1, column=col).value
            # Componi la data solo se entrambi sono presenti
            if giorno_mese and anno:
                data = f"{giorno_mese}/{anno}"
            else:
                data = ""
            results.append({
                "modello": modello,
                "header": "X",
                "data": data,
                "valore": value
            })


# Esporta in CSV
df = pd.DataFrame(results)
```

```python
import openpyxl
import pandas as pd
from datetime import datetime

def parse_date_ita(val):
    """
    Converte qualsiasi valore in stringa 'gg/mm/aaaa', se possibile.
    """
    if val is None or val == "":
        return ""
    if isinstance(val, datetime):
        return val.strftime('%d/%m/%Y')
    if isinstance(val, float):  # Data seriale Excel
        try:
            dt = datetime.fromordinal(datetime(1899, 12, 30).toordinal() + int(val))
            return dt.strftime('%d/%m/%Y')
        except Exception:
            pass
    try:
        dt = pd.to_datetime(val, errors='coerce', dayfirst=True)
        if pd.notna(dt):
            return dt.strftime('%d/%m/%Y')
    except Exception:
        pass
    # Ultima spiaggia: se già in formato "gg/mm/aaaa", la restituisce così
    return str(val)

def parse_midea(file):
    wb = openpyxl.load_workbook(file, data_only=True)
    ws = wb.active
    results = []
    start_col = 9
    max_col = ws.max_column
    header_row = 1
    anagrafica_row = 54
    start_data_row = 2
    end_data_row = 50

    for col in range(start_col, max_col + 1):
        header_value = ws.cell(row=header_row, column=col).value
        anagrafica_value = ws.cell(row=anagrafica_row, column=col).value
        if header_value is None and anagrafica_value is None:
            continue
        for row in range(start_data_row, end_data_row + 1):
            cell_value = ws.cell(row=row, column=col).value
            if cell_value is not None and str(cell_value).strip() != "" and str(cell_value).strip() != "0":
                date_value = ws.cell(row=row, column=3).value
                results.append({
                    "modello": anagrafica_value,
                    "header": header_value,
                    "data": parse_date_ita(date_value),
                    "valore": cell_value
                })
    return results

def parse_hisense(file):
    wb = openpyxl.load_workbook(file, data_only=True)
    ws = wb.active
    results = []
    max_row = ws.max_row
    max_col = ws.max_column
    start_data_col = 6
    start_model_row = 5

    for row in range(start_model_row, max_row + 1):
        modello = ws.cell(row=row, column=2).value
        if not modello:
            continue
        for col in range(start_data_col, max_col + 1):
            value = ws.cell(row=row, column=col).value
            if value is not None and str(value).strip() != "" and str(value).strip() != "0":
                header = ws.cell(row=4, column=col).value
                data = ws.cell(row=2, column=col).value
                results.append({
                    "modello": modello,
                    "header": header,
                    "data": parse_date_ita(data),
                    "valore": value
                })
    return results

def parse_homa(file):
    wb = openpyxl.load_workbook(file, data_only=True)
    ws = wb.active
    results = []
    start_data_col = 8
    end_data_col = 92
    start_model_row = 4
    max_row = ws.max_row

    for row in range(start_model_row, max_row + 1):
        modello = ws.cell(row=row, column=2).value
        if not modello:
            continue
        for col in range(start_data_col, end_data_col + 1):
            value = ws.cell(row=row, column=col).value
            if value is not None and str(value).strip() != "" and str(value).strip() != "0":
                giorno_mese = ws.cell(row=3, column=col).value
                anno = ws.cell(row=1, column=col).value
                if giorno_mese and anno:
                    data_raw = f"{giorno_mese}/{anno}"
                else:
                    data_raw = ""
                results.append({
                    "modello": modello,
                    "header": "X",
                    "data": parse_date_ita(data_raw),
                    "valore": value
                })
    return results

# ---- PATH DEI FILE ----
files_and_parsers = [
    ('../MIDEA/base spareorder MWO MIDEA_PO 2025-2 (all brands).xlsx', parse_midea),
    ('../HISENSE/base_spareorder HISENSE LVB (all brand) 2025.3.xlsx', parse_hisense),
    ('../HOMA/new Generale ordini frigo HOMA in progress_25.xlsx', parse_homa)
]

# ---- ELABORAZIONE UNIFICATA ----
all_results = []
for file_path, parser in files_and_parsers:
    data = parser(file_path)
    for row in data:
        row["source_file"] = file_path
    all_results.extend(data)

# ---- CREAZIONE DATAFRAME UNICO ----
df = pd.DataFrame(all_results)
print(df.head())
df.to_excel("output_spareorder.xlsx", index=False)
```
